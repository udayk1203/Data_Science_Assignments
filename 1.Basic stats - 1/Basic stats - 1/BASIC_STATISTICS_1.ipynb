# **Boxplots**

### Create boxplots for numerical variables to identify outliers and the interquartile range.


```python
def analyze_numerical_data(numerical_data):
    plt.figure(figsize=(25, 35))
    for i, column in enumerate(numerical_data.columns, 1):
        plt.subplot(len(numerical_data.columns), 1, i)
        sns.boxplot(x=numerical_data[column])
        plt.title(f'Boxplot of {column}')
        plt.xlabel(column)
plt.tight_layout()
plt.show()
```

```python
analyze_numerical_data(numerical_data)
```

# **Create bar charts to visualize the frequency or count of each category.**


```python
def plot_categorical_distribution(df, categorical_columns):
    for column in categorical_columns:
        plt.figure(figsize=(35, 15))
        sns.countplot(data=df, x=column, palette='bright')
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
```

# **Standardize the data**

```python
scaler = StandardScaler()
```

```python
df_standardized = scaler.fit_transform(df[numerical_columns])
df_standardized
```

```python
df_standardized = pd.DataFrame(df_standardized, columns=numerical_columns)
df_standardized
```

## **Calculate mean and standard deviation after standardization**

```python
mean_after = df_standardized.mean()
mean_after
```

```python
std_after = df_standardized.std()
std_after
```

## **Plot histograms of original data and standardized data**

```python
plt.figure(figsize=(20, 12))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[col], kde=True, color='blue', label='Original')
    plt.axvline(mean_before[col], color='r', linestyle='dashed', linewidth=2, label='Mean')
    plt.axvline(mean_before[col] + std_before[col], color='g', linestyle='dashed', linewidth=2, label='Std Dev')
    plt.axvline(mean_before[col] - std_before[col], color='g', linestyle='dashed', linewidth=2)
    plt.title(f'Original Distribution of {col}')
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 12))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i )
    sns.histplot(df_standardized[col], kde=True, color='orange', label='Standardized')
    plt.axvline(mean_after[col], color='r', linestyle='dashed', linewidth=2, label='Mean')
    plt.axvline(mean_after[col] + std_after[col], color='g', linestyle='dashed', linewidth=2, label='Std Dev')
    plt.axvline(mean_after[col] - std_after[col], color='g', linestyle='dashed', linewidth=2)
    plt.title(f'Standardized Distribution of {col}')
    plt.legend()
plt.tight_layout()
plt.show()
```
